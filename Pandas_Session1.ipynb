```python
def map_(col):
    print(col , type(col))
    if col == 'Riders':
        col = 1
    elif col == 'Devils':
        col = 2
    elif col ==  'Royals':
        col = 3
    else:
        col = 4
    return col



df['Team_Map1'] =  df.Team.apply(map_)

```
